```python
#BM25 쓸 때 토큰화 잘 하는 거 중요, BERT이용
from transformers import AutoTokenizer
from langchain.schema import Document
# BERT 다국어 토크나이저
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")

def tokenize_mixed_language_with_bert(text):
    return " ".join(tokenizer.tokenize(text))

# 분할된 문서를 BERT로 토큰화
tokenized_splits = [{"page_content": tokenize_mixed_language_with_bert(doc.page_content), "metadata": doc.metadata} for doc in splits]
# Document 객체로 변환
documents_tokenized = [Document(page_content=doc["page_content"], metadata=doc["metadata"]) for doc in tokenized_splits]
```

```python
from langchain.retrievers import EnsembleRetriever
from langchain.retrievers import BM25Retriever

# BM25 리트리버 생성 (Sparse Retriever)
bm25_retriever=BM25Retriever.from_documents(documents_tokenized)
bm25_retriever.k =  10
# FAISS 리트리버 생성 (Dense Retriever)
faiss_retriever=vectorstore_faiss.as_retriever(search_kwargs={'k':20})

# 앙상블 리트리버 생성
ensemble_retriever = EnsembleRetriever(
            retrievers=[bm25_retriever, faiss_retriever],
            weights=[0.4, 0.6]  # 가중치 설정 (가중치의 합은 1.0)
        )
```

```python
import re

def extract_answer(response):
    """
    extracts the answer from the response using a regular expression.
    expected format: "[ANSWER]: (A) convolutional networks"

    if there are any answers formatted like the format, it returns None.
    """
    pattern = r"\[ANSWER\]:\s*\((A|B|C|D|E)\)"  # Regular expression to capture the answer letter and text
    match = re.search(pattern, response)

    if match:
        return match.group(1) # Extract the letter inside parentheses (e.g., A)
    else:
        return extract_again(response)

def extract_again(response):
    pattern = r"\b[A-J]\b(?!.*\b[A-J]\b)"
    match = re.search(pattern, response)
    if match:
        return match.group(0)
    else:
        return None
```
